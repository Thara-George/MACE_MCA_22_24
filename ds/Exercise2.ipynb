```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the Iris dataset from a CSV file
csv_file = "/content/iris.csv"  # Replace with the path to your CSV file
df = pd.read_csv(csv_file)

# List of feature names (assuming the columns are named accordingly)
feature_names = df.columns[:-1]

# Create histograms for each feature
for feature in feature_names:
    plt.figure(figsize=(6, 4))
    plt.hist(df[feature], bins=20, color='skyblue', alpha=0.7)
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {feature}')
    plt.show()
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, palette='Set2')
plt.title('Box Plot of Iris Dataset Features')
plt.xlabel('Features')
plt.ylabel('Value')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
for i, feature in enumerate(feature_names):
    plt.subplot(2, 2, i + 1)  # Create a 2x2 grid of subplots
    sns.histplot(df[feature], kde=True, color='skyblue')
    plt.title(f'Distribution of {feature}')
    plt.xlabel('Value')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
sns.set(style="ticks")  # Set the style of the scatterplots
sns.pairplot(df, hue=None, markers="o", diag_kind="kde", diag_kws={"shade": True})

# Add a legend (optional)
plt.legend(loc="best")

# Show the scatterplots
plt.show()
```

```python
sns.set(style="ticks")
sns.pairplot(df, diag_kind="kde", diag_kws={"fill": True})

# Show the scatterplots
plt.show()
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(df[feature_names], alpha=0.8, figsize=(10, 10), diagonal='hist', marker='o')

# Add axis labels
plt.suptitle("Scatter Matrix for Iris Data", fontsize=16)

# Show the scatter matrix
plt.show()
```

```python
sepal_length = df[4]
sepal_width = df[4]
petal_length = df[4]
plt.figure(figsize=(10, 6))
plt.scatter(sepal_length, sepal_width, s=100 * petal_length, c='b', alpha=0.5, label='Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Bubble Chart for Iris Dataset (Sepal Length vs. Sepal Width vs. Petal Length)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from pandas.plotting import parallel_coordinates

# Create a parallel coordinate plot
plt.figure(figsize=(10, 6))
parallel_coordinates(df, 'species', colormap='viridis', alpha=0.7)

# Add a legend
plt.legend(loc='upper right')

# Add a title
plt.title('Parallel Coordinate Plot for Iris Dataset')

# Show the plot
plt.show()
```

```python
from pandas.plotting import parallel_coordinates

# Create a parallel coordinate plot without 'species'
plt.figure(figsize=(10, 6))
parallel_coordinates(df, colormap='viridis', alpha=0.7)

# Add a title
plt.title('Parallel Coordinate Plot for Iris Dataset (Without Species)')

# Show the plot
plt.show()
```

```python
sns.set(style="whitegrid")

# Create deviation charts for sepal length and petal length
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='sepal length (cm)', data=df, palette='viridis')
plt.title('Deviation Chart for Sepal Length by Species')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='petal length (cm)', data=df, palette='viridis')
plt.title('Deviation Chart for Petal Length by Species')
plt.show()
```

```python

# Set the style for the density chart
sns.set(style="whitegrid")

# Create density plots for each feature
plt.figure(figsize=(12, 8))
for i, feature in enumerate(feature_names):
    plt.subplot(2, 2, i + 1)
    sns.kdeplot(df[feature], color='b', shade=True)
    plt.xlabel(feature)
    plt.ylabel('Density')
    plt.title(f'Density Chart of {feature}')

plt.tight_layout()
plt.show()
```
